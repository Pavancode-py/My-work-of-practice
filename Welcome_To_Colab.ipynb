```
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```
data=pd.read_csv("/content/Last quater FCR. Sample.csv")
print(data)
data.describe()
```

```
x1=data["CHAT"]
y=data["AHT"]
```

```
plt.scatter(x1,y)
plt.xlabel("CHAT",fontsize=10)
plt.ylabel("AHT",fontsize=10)
plt.show()
```

```
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```
plt.scatter(x1,y,color="purple",marker="*")
# yhat=b0+b1x1
yhat=0.1880*x1+ 8.3263
plt.plot(x1,yhat,lw=5,color="orange",label="regression line")
plt.xlabel("CHAT",fontsize=10)
plt.ylabel("AHT",fontsize=10)
plt.show()
```

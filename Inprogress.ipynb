CREATING A LOGISTIC REGRESSION TO PREDICT ABSENTISM


```
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

LOAD THE PREPROCESSED DATA

```
data_preprocessed = pd.read_csv("/content/Absenteeism_preprocessed.csv")
data_preprocessed.head()
```

CREATE THE TARGETS

```
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] >
                   data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
```

```
targets
```

```
data_preprocessed['Excessive Absenteeism'] = targets
```

```
data_preprocessed.head()
```

A comment on Targets

```
targets.sum()/targets.shape[0]
```

```
data_with_targets =  data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
data_with_targets
```

SELECT THE INPUTS FOR REGRESSION

```
data_with_targets.shape
```

```
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

STANDARDIZE THE DATA

```
absentism_scaler = StandardScaler()
```

```
absentism_scaler.fit(unscaled_inputs)
```

```
scaled_inputs = absentism_scaler.transform(unscaled_inputs)
```

```
scaled_inputs.shape
```

```
scaled_inputs
```

```
# train_test_split(scaled_inputs,targets)
```

```
X_train,X_temp,y_train,y_temp = train_test_split(scaled_inputs,targets,train_size=0.8,random_state=0)
```

```
print(X_train.shape, y_train.shape)
```

```
print(X_temp.shape, y_temp.shape)
```

LOGISTIC REGRESSION

```
reg = LogisticRegression()
```

```
reg.fit(X_train,y_train)
```

```
reg.score(X_train,y_train)
```

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
sns.set()
```

```
data=pd.read_csv("/content/1.01.+Simple+linear+regression (1).csv")
data.head()
data.describe()
```

```
x=data["SAT"]
y=data["GPA"]
```

```
# check shapes
x.shape
```

```
y.shape
```

```
x_matrix= x.values.reshape(-1,1)
x_matrix.shape
# x_matrix
```

```
reg=LinearRegression()
reg.fit(x_matrix,y)
```

```
# R-Squared

reg.score(x_matrix,y)
```

```
# coefficient
reg.coef_
```

```
# contant or Intercept

reg.intercept_
```

```
# Making predictions
reg.predict([[1740]])
```

```
new_data=pd.DataFrame(data=[1740,1760,1780,1800,1801],columns=["SAT"])
new_data
```

```
reg.predict(new_data)
```

```
new_data['predicated_GPA']=reg.predict(new_data)
new_data
```

```
plt.scatter(x,y,marker="*")
# yhat = b0 + b1x1
yhat= reg.coef_*x_matrix + reg.intercept_
fig = plt.plot(x,yhat,lw=4, color='green',label="Regression line")
plt.xlabel("SAT", fontsize=10)
plt.ylabel("GPA", fontsize=10)

```

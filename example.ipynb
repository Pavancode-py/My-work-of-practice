```
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
from sklearn import linear_model
```

```
raw_data=pd.read_csv("/content/1.03.+Dummies.csv")
raw_data
```

```
data=raw_data.copy()
data["Attendance"]=data["Attendance"].map({"Yes":1,"No":0})
data
```

```
data.describe()
```

```
y=data["GPA"]
x1=data[["SAT","Attendance"]]
```

```
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```
plt.scatter(data["SAT"],y,color="purple",marker="*")
# yhat=b0+b1x1
yhat_no= 0.6439+ 0.0014*data["SAT"]
yhat_yes= 0.8665 +  0.0014*data["SAT"]
yhat= 0.0017*data["SAT"]+0.275
plt.plot(data["SAT"],yhat,lw=4,color="Blue",label="regression line")
plt.plot(data["SAT"],yhat_no,lw=4,color="Red",label="regression line1")
plt.plot(data["SAT"],yhat_yes,lw=4,color="Green",label="regression line2")
plt.xlabel("SAT",fontsize=10)
plt.ylabel("GPA",fontsize=10)
```
